```python
#fas-정재환
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import numpy as np
import csv
import json
import seaborn as sns
%matplotlib inline 
```

```python
#약 100만개의 데이터가 저장된 csv파일에서 5만개의 데이터만 가져온다
data_health = pd.read_csv("2017국민건강검진.csv", index_col=0, nrows=50000)
data_health.reset_index(drop=True, inplace=True)
data_health
```

```python
# '연령' 컬럼의 모든 열에 5를 곱한다
data_health['연령'] = data_health['연령'] * 5
```

```python
data_health
```

```python
#한글 폰트 적용 - 한글깨짐 방지
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 연령별 인원수를 막대 그래프로 표현
sns.countplot(x='연령', data=data_health)
plt.show()
```

```python
# 연령별 인원수를 세로 그래프로 표현
sns.countplot(y='연령', data=data_health)
plt.show()
```

```python
# 파이 그래프 - 각 연령별 분포 비율
fig, axes = plt.subplots(figsize=(12,6))
sizes = data_health['연령'].value_counts()
axes.pie(x = sizes, labels = sizes.index, autopct='%.1f%%')
axes.legend(title='연령')
axes.set_title("연령별 분포 비율")
plt.show()
```

```python
# 파이 그래프 - 각 연령별 분포 비율
# bbox_to_anchor를 이용해 그래프와 범주가 겹치지 않도록 설정
fig, axes = plt.subplots(figsize=(12,6))
sizes = data_health['연령'].value_counts()
axes.pie(x = sizes, labels = sizes.index, autopct='%.1f%%')
axes.legend(title='연령', loc='center right', bbox_to_anchor=(1.2, 0.5))
axes.set_title("연령별 분포 비율")
plt.show()
```

```python
# 파이 그래프 - 각 연령별 분포 비율
# bbox_to_anchor를 이용해 그래프와 범주가 겹치지 않도록 설정
fig, axes = plt.subplots(figsize=(12,6))
sizes = data_health['연령'].value_counts()
axes.pie(x = sizes, labels = sizes.index, autopct='%.1f%%')
axes.legend(title='연령', loc='center right', bbox_to_anchor=(1.8, 0.5))
axes.set_title("연령별 분포 비율")
plt.show()
```
